## Key would be Country and value will be country name

```python
from pyspark import SparkContext
import re
```

```python
class Util():
    Comma_delim = re.compile(''',(?=(?:[^"]*"[^"]*")*[^"]*$)''')
    
```

###### look for what does items do, and collectAsMap

```python
sc = SparkContext.getOrCreate()
airports = sc.textFile("in/airports.text")

country_location_airport = airports.map(lambda airport:\
                                       (Util.Comma_delim.split(airport)[3],Util.Comma_delim.split(airport)[1])
                                       )
group_airport = country_location_airport.groupByKey()
for country, tuple_air in group_airport.collectAsMap().items():
    print("{},{}".format(country,list(tuple_air)))
```
